As expected, I can see that the columns which had a lot of outliers have the most skewed distribution of values, e.g., rx, rx_with_rebates, prof, etc. However, looking at these plots I can't conclude row-wise outliers; they only give some sense of the range of values of each column and what value could potentially be treated as an outlier.

now, I will try one last approach toward the uni-variate outlier detection part of this assignment. I will standardize the value of each column, i.e., calculate its z-score. I'll then try to threshold the values further than 3 standard deviations from the mean so that they are considered outliers.

```python
from scipy import stats

z = np.abs(stats.zscore(sfrs[financial_columns]))

print(np.where(z > 3))
```

```python
sfrs_outliers = sfrs[financial_columns][(z > 3).any(axis=1)]
sfrs.iloc[sfrs_outliers.index]
```

```python
outliers_row_indices = np.where(z > 3)[0]
unique_values, counts = np.unique(outliers_row_indices, return_counts=True)
outlier_counts = np.array((unique_values, counts)).T

# changing the threshold (number of columns that are considered outliers)
threshold = 7

outliers = [x for x in outlier_counts if x[1] > threshold]

indices = [i[0] for i in outliers]
sfrs.iloc[indices]
```

Multi-Variate Outlier Detection
In this section, I will apply machine learning techniques to detect outliers. I will train some models that take all financial columns as input and output some kind of classification or categorization of whether the input is an outlier or not.

The sklearn library has implemented some of the most widely known and used machine learning algorithms for outlier detection: isolation forest and local outlier factor.


```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
isolation_forest = IsolationForest()

isolation_forest = isolation_forest.fit(sfrs[financial_columns])

inference = isolation_forest.predict(sfrs[financial_columns])


indices_if = np.where(inference == -1)[0]  # np.where() returns a tuple of one element

print("There are", len(indices_if), "members marked as outliers")

sfrs.iloc[indices_if]
```

```python
local_outlier_factor = LocalOutlierFactor()
inference = local_outlier_factor.fit_predict(sfrs[financial_columns])
indices_lof = np.where(inference == -1)[0]

print("There are", len(indices_lof), "members marked as outliers")

sfrs.iloc[indices_lof]
```

There's quite some discrepancy between the predictions of both algorithms. To become more confident in our predictions for an outlier, we can mark as such only those individuals that have been predicted as outliers by both algorithms.

```python
common_indices = list(set(indices_if).intersection(set(indices_lof)))
sfrs.iloc[common_indices]
```

111 members are identified as outliers by both algorithms. As a final presentation, I am plotting the distribution of the values in the financial columns only for the detected outliers and try to compare them with the previous one over the entire data set.

```python
_ = sfrs.iloc[common_indices][financial_columns].hist(figsize=(19, 13))
```
